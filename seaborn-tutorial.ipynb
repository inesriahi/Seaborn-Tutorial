## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Datasets

```python
flights = sns.load_dataset("flights")
flights
```

```python
flights_feb = flights[flights['month'] == 'Feb']
flights_feb
```

## Line plot

```python
sns.set_style("darkgrid")
sns.lineplot(x='year' , y='passengers', data=flights_feb)
plt.title("Year vs. Passengers")
plt.xlabel('Year')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='year' , y='passengers', data=flights,size="month", hue="month")
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue='event', style='region')
```

```python
sns.relplot(data=fmri, x="timepoint", y="signal", hue='event', style='region', kind='line', col='region', row='event')
```

## Scatter Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="smoker")
```

```python
sns.set_style('darkgrid')
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", size="size", sizes=(20, 100))
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", size="size", sizes=(20, 100), palette='plasma')
```

```python
sns.set_style('whitegrid')
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="size", size="size", sizes=(20, 100), palette='Purples')
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time",style="time", markers={"Lunch":"s", "Dinner":"D"})
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="time", style="sex", kind="scatter", col="sex", row="day")
```
