## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Datasets

```python
flights = sns.load_dataset("flights")
flights
```

```python
flights_feb = flights[flights['month'] == 'Feb']
flights_feb
```

## Line plot

```python
sns.set_style("darkgrid")
sns.lineplot(x='year' , y='passengers', data=flights_feb)
plt.title("Year vs. Passengers")
plt.xlabel('Year')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='year' , y='passengers', data=flights,size="month", hue="month")
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue='event', style='region')
```

```python
sns.relplot(data=fmri, x="timepoint", y="signal", hue='event', style='region', kind='line', col='region', row='event')
```

# Regression Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
tips['size'].value_counts()
```

```python
sns.regplot(x='size', y='tip', data=tips)
```

```python
sns.regplot(x='size', y='tip', data=tips, x_jitter=0.1)
```

```python
sns.regplot(x='size', y='tip', data=tips, x_estimator=np.median)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue="smoker", col="smoker", row="sex", truncate=False)
```

```python
tips["smoker"] = tips["smoker"] == 'Yes'
```

```python
tips
```

```python
sns.lmplot(x='total_bill', y='smoker', data=tips, logistic=True)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue="smoker", col="smoker", 
           scatter_kws={'s': 100, 'edgecolor':'black'}, 
          line_kws={'lw': 6, 'ls': '--'})
```
